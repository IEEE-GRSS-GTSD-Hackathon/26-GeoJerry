```python
import leafmap
```

```python
#MAXAR DATA
leafmap.maxar_collections()
```

```python
collection = 'Kahramanmaras-turkey-earthquake-23'
url = leafmap.maxar_collection_url(collection, dtype='geojson')
url
```

```python
import geopandas as gpd
df = gpd.read_file(url)
print(f'Total number of images: {len(df)}')
df.head()
```

```python
map = leafmap.Map()
map.add_gdf(df, layer_name='Footprints')
map
```

```python
pre_gdf = leafmap.maxar_search(collection, end_date='2023-02-06')
print(f'Pre disaster count: {len(pre_gdf)}')
pre_gdf.head()
```

```python
post_gdf = leafmap.maxar_search(collection, start_date='2023-02-06')
print(f'Post disaster count {len(post_gdf)}')
post_gdf.head()
```

```python
#Defining the coordinate
coo = m.user_roi_bounds()
if coo is None:
    coo = [36.8715, 37.5497, 36.9814, 37.6019]
```

```python
pre_disaster = leafmap.maxar_search(collection, bbox=coo, end_date='2023-02-06')
pre_disaster.head()
```

```python

post_disaster = leafmap.maxar_search(collection, bbox=coo, start_date='2023-02-06')
post_disaster.head()
```

```python
pre_dt = pre_disaster['catalog_id'].values[0]
post_dt = post_disaster['catalog_id'].values[0]
```

```python
pre_stac = leafmap.maxar_tile_url(collection, pre_dt, dtype='json')
```

```python
post_stac = leafmap.maxar_tile_url(collection, post_dt, dtype='json')
```

```python
import leafmap.foliumap as leafmap
m = leafmap.Map()
m.split_map(
    left_layer=pre_stac,
    right_layer=post_stac,
    left_label='Pre-disaster',
    right_label='Post-disaster',
)
m.set_center(36.9265, 37.5762, 16)
m
```

```python
pre_images = pre_disaster['visual'].tolist()
post_images = post_disaster['visual'].tolist()
```

```python
leafmap.maxar_download(pre_images)
```

```python
leafmap.maxar_download(post_images)
```
